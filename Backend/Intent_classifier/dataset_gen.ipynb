```python
import random

names = ["jefin", "ravi", "anu", "mom", "dad", "priya", "arjun"]
amounts = ["100", "250", "500", "750", "1000", "1500", "2000", "5000"]
phones = ["9912381230", "9876543210", "9123456789"]
upis = ["jefin upi", "ravi okhdfc", "anu ybl", "sam icici"]

transfer_templates = [
    # Basic Commands
    "send {amount} to {name}",
    "pay {amount} to {name}",
    "transfer {amount} to {name}",
    "give {name} {amount} rupees",
    "{amount} to {name}",
    "debit {amount} and credit to {name}",

    # Conversational Requests
    "can you please transfer {amount} to {name}",
    "i want to send {amount} to {name}",
    "i need to make a payment of {amount} to {name}",
    "i'd like to give {name} {amount}",
    "could you send {amount} to {name} for me",
    "alright, let's pay {name} {amount}",
    "please send {name} {amount}",
    "i have to pay {name} {amount}",
    "it's time to pay {name} {amount}",

    # Implicit Amount
    "send money to {name}",
    "pay {name}",
    "make a payment to {name}",
    "i need to send money to {name}",
    "transfer to {name}",
    
    # Implicit Recipient
    "send {amount}",
    "pay {amount}",
    "i need to transfer {amount} rupees",
    "make a payment of {amount}",

    # Implicit Amount & Recipient
    "i want to send money",
    "i need to make a payment",
    "start a new transaction",
    "money transfer",

    # Phone Number Variations
    "pay {amount} to this number {phone}",
    "send {amount} to mobile number {phone}",
    "transfer {amount} to the phone number {phone}",
    "pay the number {phone} {amount}",
    "on this number {phone}, send {amount}",
    "i need to send money to {phone}",
    "pay this number {phone}",
    
    # UPI ID Variations
    "pay {amount} to this upi id {upi}",
    "send {amount} to upi id {upi}",
    "transfer {amount} to {upi}",
    "pay the upi address {upi} an amount of {amount}",
    "i want to pay {upi} {amount}",
    "upi payment of {amount} to {upi}",
    "pay via upi to {upi}",
    "send money to {upi}",

    # Alternative Phrasing & Structure
    "initiate a transfer of {amount} to {name}",
    "authorize a payment of {amount} to {name}",
    "new payment to {name} for {amount}",
    "for {name}, please send {amount}",
    "credit {amount} to {name}'s account",
    "remit {amount} to {name}",
    "wire {amount} to {name}",
    "my payment to {name} is {amount}",
    "use upi to send {amount} to {name}",
    "debit my account for {amount} and send to {name}",
    "the amount is {amount}, send it to {name}",

    "i need to transfer {amount} to {name} via upi",
    "send {amount} to {name} using upi",        
    "i want to pay {name} {amount} through upi",
    "make a upi payment of {amount} to {name}",
    "i'd like to transfer {amount} to {name} using upi",
    "send {amount} to {name} through upi",
    "i need to pay {name} {amount} via upi",
    "transfer {amount} to {name} using upi",
    "i want to send {amount} to {name} via upi",

    # More conversational/casual
    "shoot {amount} over to {name}",
    "zap {amount} to {upi}",
    "send some money to {name}, say {amount}",
    "let's get this payment done, {amount} to {name}",
    "ok, send {amount} to {name}",
    "alright, pay {name} {amount}",
    "to {name}, send {amount}",
    "to this number {phone}, pay {amount}",
    "for {upi}, transfer {amount}",
    "this upi id {upi}, pay {amount} rupees",
    "payment to {name}, {amount}"
]
request_templates = [
    # Basic Commands
    "request {amount} from {name}",
    "ask {name} for {amount}",
    "get {amount} from {name}",
    "collect {amount} from {name}",
    "send a payment request to {name} for {amount}",

    # Conversational Requests
    "can you get {amount} from {name}",
    "i need to request {amount} from {name}",
    "please ask {name} for {amount} rupees",
    "could you please request {amount} from {name}",
    "i want to collect {amount} from {name}",
    "i'd like to ask {name} for {amount}",
    "let's ask {name} for {amount}",
    
    # Implicit Amount
    "request money from {name}",
    "ask for money from {name}",
    "get money from {name}",
    "send a payment request to {name}",
    "collect money from {name}",
    "i need money from {name}",
    "ping {name} for payment",
    
    # Implicit Requester
    "request for money",
    "i need to ask for money",
    "start a new money request",

    # Alternative Phrasing
    "initiate a money request for {amount} from {name}",
    "create a request for {amount} to be paid by {name}",
    "tell {name} to send me {amount}",
    "{name} owes me {amount}, send a request",
    "generate a payment request for {name} of {amount}",
    "bill {name} for {amount}",
    "invoice {name} for {amount}",
    "remind {name} to send me {amount}",
    "i'm expecting {amount} from {name}, send a request",
    "i need to get paid {amount} by {name}",
    "send {name} a bill for {amount}",
    
    # Phone/UPI Variations
    "request {amount} from the number {phone}",
    "ask for {amount} from upi id {upi}",
    "send a payment request for {amount} to the phone number {phone}",
    "get {amount} from {phone}",
    "request money from upi handle {upi}",
    "ask for {amount} from {upi}",

    # Hinglish Variations
    "{name} se {amount} maango",
    "{name} se {amount} lene hai",
    "{name} se {amount} request karo",
    "request karo {amount} {name} se",
    "{name} se paise maang lo",
    "{phone} number se {amount} maango",
    "is upi id {upi} se {amount} request karo",

    # More conversational/casual
    "ping {name} for {amount}",
    "nudge {name} for {amount}",
    "send {name} a reminder for {amount}",
    "let's get {amount} from {name}",
    "alright, ask {name} to pay {amount}",
    "ok, request {amount} from {name}",
    "from {name}, I need {amount}",
    "money request from {name} for {amount}",
    "{name}, ask for {amount}",
    "from {name}, please get {amount}",
    "it's a request for {amount} from {name}",
    "you need to ask {name} to pay me {amount}",
    "send a due notification of {amount} to {name}",
    "can you ping {name} for a payment of {amount}",
    "formally request {amount} from {name}",
    "i need to invoice {name} for {amount}",
    "the amount is {amount}, request it from {name}",
    "request a payment of {amount} from {name}",
    "i am owed {amount} by {name}, please request it",
    "send a upi request to {name} for {amount}",
    "request via upi from {name} an amount of {amount}",
    "start a request for {amount} from {name}",
    "{amount} from {name}, please request"
]

balance_templates = [
    # Basic Questions
    "what is my current balance",
    "check account balance",
    "how much money do i have",
    "show me my balance",
    "balance enquiry",
    "what's my balance",
    "tell me my account balance",

    # Conversational Requests
    "can you check my balance",
    "i want to know my balance",
    "i need to check my balance",
    "i'd like to check my balance",
    "could you fetch my account balance",
    "please provide my account balance",
    "let me know my balance",

    # Short/Direct Commands
    "available balance",
    "balance please",
    "my balance",
    "get my balance",
    "current balance",
    "check funds",
    "balance check please",
    "balance status",
    "latest balance",
    
    # Indirect Queries
    "do i have enough money",
    "is my balance sufficient",
    "how much can i spend",
    "do i have money for a {amount} purchase",
    "can i afford {amount}",
    "check if i can pay {amount}",

    # Alternative Phrasing
    "how much do i have in my account",
    "what are my available funds",
    "what is the balance in my account",
    "how much funds are available",
    "what's the total amount in my account",
    "verify my account balance",
    "fetch my current balance",
    "display my balance",
    "what is the available limit",
    "i need a balance update",
    "give me a balance report",
    "what is my remaining balance",
    "check upi balance",

    # Hinglish Variations
    "balance kitna hai",
    "mere account me kitne paise hain",
    "paise check karo",
    "account ka balance batao",
    "kitna balance hai",
    "paise kitne bache hain",

    # Casual/Slang
    "show me the money",
    "am i broke",
    "how's my account looking",
    "what's in the bank",
    "what's the damage",
    "tell me the numbers",
    "how much money is left",
    "i wanna check my balance",

    # Formal Phrasing
    "i would like to make a balance enquiry",
    "provide balance details",
    "what's my financial status",
    "financial overview",

    # More conversational
    "let's see my balance",
    "ok check my balance now",
    "alright show me my balance",
    "what is my upi account balance",
    "check my savings balance",
    "is there enough money in my account",
    "reveal my balance"
]
def generate_samples(intent, templates, count):
    samples = []
    for _ in range(count):
        template = random.choice(templates)
        sample = template.format(
            amount=random.choice(amounts),
            name=random.choice(names),
            phone=random.choice(phones),
            upi=random.choice(upis)
        )
        samples.append((sample, intent))
    return samples

data = []
data += generate_samples("transfer_money", transfer_templates, 300)
data += generate_samples("request_money", request_templates, 300)
data += generate_samples("check_balance", balance_templates, 300)
random.shuffle(data)

with open("voice_upi_dataset.csv", "w") as f:
    f.write("text intent\n")
    for text, label in data:
        f.write(f"{text}, {label}\n")
```
